```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dotss = sns.load_dataset('dots')
dotss.head()
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



# defining a color palette
p = sns.color_palette('rocket_r')

# plot lineplot

sns.relplot(
    data = dotss,
    x='time',y = 'firing_rate',
    hue = 'coherence',size='choice',col='align',
    kind='line', size_order= ["T1","T2"], palette=p,
    height=5, aspect=.75, facet_kws = dict(sharex=False)
)
```

```python
# load dataset of tips
tips = sns.load_dataset('tips')

# Draw graph
sns.boxplot(x='day',y='total_bill',hue='smoker',palette=['m','g'],data=tips)
sns.despine(offset=10,trim=True)
```

```python
tips.head()
```

```python
# Draw violin plot by using tips dataset
sns.set_theme(style='whitegrid')
sns.violinplot(data=tips, x='day',y='total_bill',hue='smoker',palette={'Yes':'b','No':'.85'},
               split=True,inner='quart', linewidth=1)
sns.despine(left=True)
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')

diamonds = sns.load_dataset('diamonds')

diamonds.head()
```

```python
# draw a scatter plot while assigning point colors and sizes to different variables in the dataset
f,ax = plt.subplots(figsize=(6.5,6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ['I1','SI2','VS2','VS1','VVS2','VVS1','IF']
sns.scatterplot(
                x='tip',y='total_bill',
                size='size',
                palette='ch:r=-.2,d=.3_r',
                hue_order=clarity_ranking,
                sizes=(1,8),linewidth=0,data=tips,ax=ax
                )
```

```python
f,ax = plt.subplots(figsize=(6.5,6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ['I1','SI2','VS2','VS1','VVS2','VVS1','IF']
sns.scatterplot(
                x='carat',y='price',
                size='depth',
                palette='ch:r=-.2,d=.3_r',
                hue_order=clarity_ranking,
                sizes=(1,8),linewidth=0,data=diamonds,ax=ax
                )
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='ticks')

# intialize the figure with logarthimic x-axis
f,ax = plt.subplots(figsize=(7,6))
ax.set_xscale('log')

# Load Dataset
kashti = sns.load_dataset('titanic')

# Plot orbital period with horizontal boxes
sns.boxplot(x='age',y='class',data=kashti, whis=[0,100],width=.6,palette='vlag')

# Add in points to show each observation 
sns.stripplot(x='age',y='class',data=kashti,size=4, linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel='')
sns.despine(trim=True,left=True)
```
