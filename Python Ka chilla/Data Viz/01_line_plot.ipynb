#  Import Libraries
Seaborn automatically installs these libraries
- numpy
- pandas
- scipy
- matplotlib

```python
#import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 

dataaa= pd.read_csv('births.csv')
# dataaa
plt.title("Birth's Plot")
sns.lineplot(x="month", y="births", data = dataaa)
plt.show()
```

## Adding Limits

```python
#import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 

dataaa= pd.read_csv('births.csv')
# dataaa
plt.title("Birth's Plot")
sns.lineplot(x="month", y="births", data = dataaa)

plt.xlim(1)
plt.ylim(8000)
plt.show()
```

```python
#import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 



dataaa= pd.read_csv('births.csv')
# dataaa
plt.title("Birth's Plot")
sns.lineplot(x="month", y="births", data = dataaa)
sns.set_style("ticks")
plt.show()
```

## Size of Figure

```python
#import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
 
dataaa= pd.read_csv('births.csv')
# dataaa

plt.figure(figsize=(8,6))
sns.lineplot(x="month", y="births", data = dataaa)
plt.title("Birth's Plot")
plt.show()
```
